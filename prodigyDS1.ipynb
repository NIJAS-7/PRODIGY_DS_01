**DATASET LOADING**

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/New folder/Life Expectancy Data.csv')
df
```


**Imported BarChart for visualizing the life expectancy of differnt countries**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(35,30))
plt.bar(df['Country'],df['Life expectancy '])
plt.xlabel('countries')
plt.ylabel('life expectancy')
plt.title('life expectancy by country')
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(rotation=90)
plt.show()
```
